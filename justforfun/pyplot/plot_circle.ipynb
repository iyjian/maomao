```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
step = 1e-3

def plot_circle(O, r):
    x = np.arange(-r, r, step)
    y = np.sqrt(r**2 - x**2)
    x += O[0]
    y += O[1]
    (np.concatenate((y, -np.flip(y))))
    plt.plot((np.concatenate((x, np.flip(x)))), (np.concatenate((y, -np.flip(y)))))
```

```python
plot_circle([0, 0], 1)
plt.show()
```

```python
x = np.arange(-1, 1, 0.001)
def f(x):
    return np.sqrt(1 - np.square(x))
```
