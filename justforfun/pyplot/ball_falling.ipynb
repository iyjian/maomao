```python
g = 9.8

def plot_dropping(h, v, step = 1e-3):
    t = np.arange(0, np.sqrt(2 * h / g), step)
    h_arr = h - (g/2) * (t**2)
    l_arr = v * t
    plt.plot(l_arr, h_arr)
    plt.show()
```

```python
plot_dropping(10, 100)
```
