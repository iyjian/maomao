this is to discibe the falling of the ball being throwed horizontally

```python
import matplotlib.pyplot as plt
import math
```

```python
g = 9.8

def h_current(t, h):
    return h - (g/2) * (t**2)

def l_current(t, v):
    return v * t

def plot_dropping(h, v, step = 0.01):
  end = math.sqrt(2 * h / g)
  result1 = []
  result2 = []
  current_time = 0
  while current_time < end:
    result1.append(h_current(current_time, h))
    result2.append(l_current(current_time, v))
    current_time += step
  plt.plot(result2, result1)
  plt.show()
    
```

```python
plot_dropping(0.1, 0.01)
```
